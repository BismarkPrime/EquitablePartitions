```python
def undirectize(G):
    G = G.to_undirected()
    G = G.to_directed()
    return G

def reset():
    G = spec.random_graph_generator(20,.1)
    G = undirectize(G)
    iterations = [0]
    perc_list = []
    DataSlave = EPData()
    DataSlave.loadFromNetworkX(G)
    perc_list.append(DataSlave.getPercentNonTrivial())
    return G, iterations, perc_list
```

```python
G, iterations, perc_list = reset()

for i in range(20):
    G = spec.singleNodeSpecialize_CompleteIn(G,spec.ev_centrality_node)  #random.choice(list(G.nodes)))
    G = undirectize(G)
    DataSlave = EPData()
    DataSlave.loadFromNetworkX(G, progress_bars=False)
    perc_list.append(DataSlave.getPercentNonTrivial())
    iterations.append(i+1)
    
```

```python
plt.plot(iterations,perc_list)
plt.show()
```
